```python
import pandas as pd
```

```python
# importing csv file containing 
# play-by-play data for NFL games (2009-2018)
pbpk = pd.read_csv('../data/pbpk.csv')
pbpk.head(5)
```

```python
# creating a special identifier column for each game to use
# for merging with armchairanalysis.com data files
pbpk['date_id']=pbpk['game_date']+pbpk['home_team']
```

```python
pbpk.shape
```

```python
# importing playing conditions file
conditions = pd.read_csv('../data/conditions.csv')
conditions.head(5)
```

```python
# adding the conditions data file to the main play by play file
pbpk_cnd = pd.merge(pbpk,conditions, how='left', on='date_id')
pbpk_cnd.head()
```

```python
# whittle down columns by eliminating the ones not needed for analysis
pbpfoc = pbpk_cnd[['play_id', 'game_id','date_id', 'home_team', 'away_team', 'posteam',
       'posteam_type', 'defteam', 'side_of_field', 'yardline_100', 'game_date',
       'quarter_seconds_remaining', 'half_seconds_remaining',
       'game_seconds_remaining', 'game_half', 'quarter_end', 'drive', 'sp',
       'qtr', 'down', 'goal_to_go', 'time', 'yrdln', 'ydstogo', 'ydsnet',
       'desc', 'play_type', 'yards_gained', 'shotgun', 'no_huddle',
       'qb_dropback','qb_scramble', 'pass_length',
       'pass_location', 'air_yards', 'yards_after_catch', 'run_location',
       'run_gap', 'home_timeouts_remaining',
       'away_timeouts_remaining', 'timeout', 'timeout_team', 'td_team',
       'posteam_timeouts_remaining', 'defteam_timeouts_remaining',
       'total_home_score', 'total_away_score', 'posteam_score',
       'defteam_score', 'score_differential', 'posteam_score_post',
       'defteam_score_post', 'score_differential_post','first_down_rush', 'first_down_pass', 'first_down_penalty',
       'third_down_converted', 'third_down_failed', 'fourth_down_converted',
       'fourth_down_failed', 'incomplete_pass', 'interception','fumble_forced', 'fumble_not_forced',
       'fumble_out_of_bounds', 'solo_tackle', 'penalty',
       'tackled_for_loss', 'fumble_lost', 'qb_hit', 'rush_attempt', 'pass_attempt',
       'sack', 'touchdown', 'pass_touchdown', 'rush_touchdown',
       'return_touchdown','stad', 'temp', 'humd', 'wspd', 'wdir', 'cond',
       'surf']]
```

```python
# saving focus data set to csv
pbpfoc.to_csv('../data/pbpfoc.csv', index=False)
```
